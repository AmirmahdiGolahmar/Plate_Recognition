##تعریف تابع تبدیل مختصات پلاک به فرمت یولو و اجرای آن روی داده های آموزش و ارزیابی

```python
def xml_to_yolo(bbox, w, h):
  x_center=((bbox[2]+bbox[0])/2)/w
  y_center=((bbox[3]+bbox[1])/2)/h
  width=(bbox[2]-bbox[0])/w
  height=(bbox[3]-bbox[1])/h
  return [x_center, y_center, width, height]
```

```python
classes = []
input_dir = '/content/drive/MyDrive/project2/test/xml_test'
output_dir = '/content/drive/MyDrive/project2/test/labels'
image_dir = '/content/drive/MyDrive/project2/test/jpg_test'

if not os.path.isdir(output_dir):
    os.mkdir(output_dir)

files = glob.glob(os.path.join(input_dir, '*.xml'))

def xml_to_yolo(pil_bbox, width, height):
    x_center = (pil_bbox[0] + pil_bbox[2]) / 2.0 / width
    y_center = (pil_bbox[1] + pil_bbox[3]) / 2.0 / height
    w = (pil_bbox[2] - pil_bbox[0]) / width
    h = (pil_bbox[3] - pil_bbox[1]) / height
    return [x_center, y_center, w, h]

for fil in files:
    basename = os.path.basename(fil)
    filename = os.path.splitext(basename)[0]
    image_path = os.path.join(image_dir, f"{filename}.jpg")

    if not os.path.exists(image_path):
        print(f"{filename} image not exist")
        continue

    result = []

    tree = ET.parse(fil)
    root = tree.getroot()

    image = Image.open(image_path)
    image_width, image_height = image.size

    for obj in root.findall('object'):
        label = obj.find('name').text

        if label == "کل ناحیه پلاک":
            if label not in classes:
                classes.append(label)
            index = classes.index(label)
            pil_bbox = [int(x.text) for x in obj.find('bndbox')]
            yolo_bbox = xml_to_yolo(pil_bbox, image_width, image_height)

            bbox_string = ' '.join([str(x) for x in yolo_bbox])
            result.append(f"{index} {bbox_string}")

    if result:
        with open(os.path.join(output_dir, f"{filename}.txt"), 'w', encoding='utf-8') as f:
            f.write('\n'.join(result))

with open('classes.txt', 'w', encoding='utf-8') as f:
    f.write(json.dumps(classes))

print("xml to yolo done")
```
